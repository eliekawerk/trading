#### 2.1 Options and their value

* American Option: right to buy or sell unit of underling during a predetermined period
* European Option: right to buy or sell unit of underlying at a predetermined date
* Payoff value of an option is the max between the strike price K and the index level S at time T
* $h_t(S,K) = max[S_t - K, 0]$
* $S_t$ = index level at time $t$
* $K$ = Strike price at time $t$
* $t$ is the maturity date of an option
* For the example below our strike is 800 and we can see that payoff is positive at index values > 800
* Out of the money: For all values of $S < K$ ($S > K$) for a call (put)
* In the money: For all values of $S > K$ ($S < K$) for a call (put)
* At the money: For all values of $S$ ~ $K$

* $K$: strike price of option
* $T$: Maturity date 
* $r$: constant riskless short-rate
* $\sigma$: volatility of index, standard deviation of index returns

#### 2.2 Vanilla vs Exotic options

* Vanilla options have liquid markets because the products are commoditized
* Exotic options are bespoke with the sellers created single-use products and are not traded
* Pricing exotic options then is a challenging and important task
* Option writers must be able to hedge their bets against they options they write
* Valuations of those options feeds into the hedging process
* __Any model used to price exotics should accurately price vanillas whose price can be verified in the market__

#### 2.3.1 Market Risks Affecting Equity Derivatives
* Price risk
* volatility risk
* jump or crash risk
* interest rate risk
* correlation risk
* liquidity risk: for some intstruments extremely important when you face jump/crash risk
* default risk: liquidity risk may affect defaults 

#### 2.3.2 Other risks
* Model risk: your model is wrong or grossly unhelpful
* Operational risk: Technology risk, maybe key people

#### 2.4 Hedging
* Main goal of hedging program is replicate hedged derivative's payoff 
* Dynamic hedging: Use options and epdate positions frequently
* Static Hedging: use options and hold constant position until maturity
* Arbitrage/Risk Neutral hedging
* Bjork (2004) Models with continuous prices
* Cont and Tankov (2004): models with process that jump
* Dynamic hedging requires frequent rebalancing and this can get expensive or unrealistic
* Market jumps would lead to the need for an excessive number of edges (almost infinite)

#### 2.5 Market based Valuation
* From institutional perspective, to maintain a business in equity derivative writing
* Market realities: is there liquidity in the underlying
* Market model: Is the model realistic wrt to the market, the underlying and the option market. 
* Vanilla instrument pricing: You must be able to price industrial quantities of vanilla instruments
* Model calibration: The model must be able to accuratel reproduce prices of liquid vanilla options
* Exotic instrument pricing: You need techniques to price exotics
* Hedging: The valuation process should enable you to hedge options for which you are the seller
